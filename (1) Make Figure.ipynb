# Introduction

In this notebook, we will make a combined figure using the significant hits from the H3N2 (human) and H3N8 (seal) viruses.

```python
import pandas as pd
from matplotlib.gridspec import GridSpec
```

```python
# Open seal data
seal_data = pd.read_csv('H3N8 Data.csv', index_col='Chart Number', na_values='#NUM!')
seal_data
```

```python
# Open human H3 data
human_data = pd.read_csv('H3N2 Data.csv', index_col='Chart Number', na_values='#NUM!')
human_data
```

The goal here is to join the two dataframes to make one master dataframe that contains the significant hits of both lists.

```python
# Prepare significant seal hits dataframe for joining
seal_data.columns = map(lambda x: x + "_Seal", seal_data.columns)
human_data.columns = map(lambda x: x + "_Human", human_data.columns)
# Perform a join on the two dataframes
joined = seal_data.sort('p-value_Seal').join(human_data.sort('p-value_Human'))
joined
```

```python
# significant = joined[(joined['p-value_Human'] < 0.01) | (joined['p-value_Seal'] < 0.01)]
# We do not want to show only the p-value < 0.01 hits. Therefore, commented out the above line.
significant = joined
significant.reset_index(inplace=True)
significant['alpha(2-3)'] = significant['Glycan Name_Seal'].str.contains('a2-3')
significant['alpha(2-6)'] = significant['Glycan Name_Seal'].str.contains('a2-6')
significant = significant.sort(['alpha(2-3)', 'alpha(2-6)'], ascending=[False, False]).reset_index()
del significant['index']
sig = pd.DataFrame()
for group in significant.groupby('alpha(2-3)'):
    if group[0] == True:
        sig = sig.append(group[1].sort('alpha(2-6)', ascending=False))
    if group[0] == False:
        sig = sig.append(group[1].sort('alpha(2-6)', ascending=True))
sig = sig.reset_index()
del sig['index']
sig.to_csv('Joined Hits.csv')
sig
```

Having inspected the CSV file in Excel, I have noticed that they are already well separated between alpha(2-3) and alpha(2-6). Now, let's make a plot.
