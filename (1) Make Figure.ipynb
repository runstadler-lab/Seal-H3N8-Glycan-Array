# Introduction

In this notebook, we will make a combined figure using the significant hits from the H3N2 (human) and H3N8 (seal) viruses.

```python
import pandas as pd
from matplotlib.gridspec import GridSpec
```

```python
# Open seal data (pre-complexed)
seal_data = pd.read_csv('H3N8 Data.csv', index_col='Chart Number', na_values='#NUM!')
seal_data
```

```python
# Open seal data (sequential)
seal_data_sequential = pd.read_csv('H3N8 Data (sequential).csv', index_col='Chart Number', na_values='#NUM!')
seal_data_sequential
```

```python
seal_data['Average RFU_sequential'] = seal_data_sequential['Average RFU']
seal_data['StDev_sequential'] = seal_data_sequential['StDev']
```

```python
# Open human H3 data (pre-complexed)
human_data = pd.read_csv('H3N2 Data.csv', index_col='Chart Number', na_values='#NUM!')
human_data
```

```python
# Open human H3 data (sequential)
human_data_sequential = pd.read_csv('H3N2 Data (sequential).csv', index_col='Chart Number', na_values='#NUM!')
human_data_sequential
```

```python
human_data['Average RFU_sequential'] = human_data_sequential['Average RFU']
human_data['StDev_sequential'] = human_data_sequential['StDev']
```

The goal here is to join the two dataframes to make one master dataframe that contains the significant hits of both lists.

```python
# Prepare significant seal hits dataframe for joining
seal_data.columns = map(lambda x: x + "_Seal", seal_data.columns)
human_data.columns = map(lambda x: x + "_Human", human_data.columns)
# Perform a join on the two dataframes
joined = seal_data.sort('p-value_Seal').join(human_data.sort('p-value_Human'))
joined
```

```python
# significant = joined[(joined['p-value_Human'] < 0.01) | (joined['p-value_Seal'] < 0.01)]
# We do not want to show only the p-value < 0.01 hits. Therefore, commented out the above line.
# significant = joined
joined.reset_index(inplace=True)
joined['alpha(2-3)'] = joined['Glycan Name_Seal'].str.contains('a2-3')
joined['alpha(2-6)'] = joined['Glycan Name_Seal'].str.contains('a2-6')
joined = joined.sort(['alpha(2-3)', 'alpha(2-6)'], ascending=[False, False]).reset_index()
del joined['index']
sorted_joined = pd.DataFrame()
for group in joined.groupby('alpha(2-3)'):
    if group[0] == True:
        sorted_joined = sorted_joined.append(group[1].sort('alpha(2-6)', ascending=False))
    if group[0] == False:
        sorted_joined = sorted_joined.append(group[1].sort('alpha(2-6)', ascending=True))
sorted_joined = sorted_joined.reset_index()
del sorted_joined['index']
sorted_joined.to_csv('Joined Hits.csv')
sorted_joined
```

Having inspected the CSV file in Excel, I have noticed that they are already well separated between alpha(2-3) and alpha(2-6). Now, let's make a plot.
